## Feature Selection

```python
movies.columns
```

```python
movies=movies[['id','title','genres','original_language','overview','popularity','production_companies','release_date']]
movies
```

```python
movies["tags"]=movies["genres"]+movies["overview"]+movies["production_companies"]+movies["original_language"]
```

```python
movies.head()
```

```python
new_data=movies.drop(columns=['genres','original_language','overview','production_companies'])

new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=4799, stop_words="english")
cv
```

```python
vector=cv.fit_transform(new_data["tags"].values.astype("U")).toarray()
```

```python
vector.shape
```

## Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
new_data[new_data["title"]=="The Dark Knight Rises"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=new_data[new_data["title"]==movies].index[0]
    distance= sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
    for i in distance [0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("John Carter")
```

```python
recommend("Iron Man")
```
